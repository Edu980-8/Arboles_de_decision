```python
import pandas as pd
import matplotlib.pyplot as plt


titanic = pd.read_csv("titanic.csv", sep=",")
titanic
```

```python
titanic.head(10)
```

```python
columns_to_drop = ['Name','Fare']
titanic.drop(columns_to_drop, axis="columns", inplace=True)
titanic
```

```python
titanic.columns = ['Survived','Pclass','Sex','Age','SibSp','Parch']
titanic.head()
```

```python
print(titanic.shape)
```

```python
titanic.dtypes
```

```python
titanic = pd.get_dummies(titanic,columns=["Sex"], drop_first= True)
titanic.dtypes
```

```python
titanic.head()
```

```python
titanic.rename(columns={'Sex_male':'Sex'}, inplace= True)
titanic = titanic[['Survived','Pclass','Sex','Age','SibSp', 'Parch']]
titanic.head()
```

```python
x = titanic.drop("Survived", axis=1)
y = titanic.Survived
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test, y_train, y_test = train_test_split(x,y, test_size = 0.30, random_state =00000)
```

```python
# Arbol de decision
from sklearn.tree import DecisionTreeClassifier
# Creacion del modelo
tree = DecisionTreeClassifier(max_depth = 2, random_state =00000 ) 

# --->No se le pone una profundidad muy grande porq con pocos datos tiende a sobreajustarse
tree.fit(X_train, y_train)
```

## Evaluacion del modelo

```python
# Calculo de las predicciones en Train y Test
y_train_pred = tree.predict(X_train)
y_test_pred = tree.predict(X_test)
```

```python
# Calculo de metricas (Como estamos en un problema de clasificacion se usa el accuracy)
from sklearn.metrics import accuracy_score
train_accuracy = accuracy_score(y_train, y_train_pred)
test_accuracy = accuracy_score(y_test, y_test_pred)
print(f'El accuracy del entrenamiento es: {train_accuracy},\n El accuracy del test es: {test_accuracy}')
# Como los valores de los accuracys son cercanos se puede inferir que el modelo es decente...
```

```python
import seaborn as sns
importances = tree.feature_importances_
columns = x.columns
sns.barplot(columns, importances, palette = 'bright', saturation = 2.0, edgecolor = 'black', linewidth=2)
plt.title('Importancia de cada Feature')
plt.show()
```

### Se puede Observar que la variable mas relevante para la clasificacion realizada por el arbol de decision es la variable sex

#### Ademas se puede ver que al tener valores  de accuracy cercanos nuestro modelo se maneja de manera adecuada
